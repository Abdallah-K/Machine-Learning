```python
import pandas as pd
```

```python
df = pd.read_csv("dataset/salaries.csv")
df.head()
```

```python
inputs = df.drop("salary_more_then_100k",axis = 1) # axis = 1 tp remove the column
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
#You're creating three separate LabelEncoder objects to convert categorical 
#text data (like company names, job titles, and degrees) into numerical form.
#LabelEncoder is designed to handle one column at a time. It learns the mapping between 
#unique values and integers based on the input it's fitted on.
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target) # score is 1 since i use the same dataset
```

```python
model.predict([[2,0,1]])
```
